##Notions générales

**Un tuple ou n-uplet est une suite ordonnée d'éléments non modifiables qui peuvent être de n'importe quel type.**

* Les éléments sont ordonnés avec un indice
* On ne peut pas modifier les éléments par affectation

###Création de tuples

Pour créer un tuple on utilise les parenthèses avec la virgule comme séparateur :

Ces instructions créent et affichent un tuple qui contient : la lettre 'a' , son code ASCII et sa représentation en binaire

```
mon_tuple1=("a",ord("a"),bin(ord("a")))
print(mon_tuple1)
```

###Accéder à un élément d'un tuple
Pour accéder à un élément du tuple, on utilise l'instruction : mon_tuple[index_de_l'élément]

```
mon_tuple1=("a",ord("a"),bin(ord("a")))
print(mon_tuple1)
print(mon_tuple1[2])
```
